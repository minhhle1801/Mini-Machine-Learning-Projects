```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Loading Data

yelp = pd.read_csv('yelp.csv')
yelp.head()
```

```python
# Best and Worst Reviews

X = yelp.loc[(yelp['stars']==1) | (yelp['stars']==5), ['text']].squeeze()
y = yelp.loc[(yelp['stars']==1) | (yelp['stars']==5), ['stars']].squeeze()
```

```python
# Train Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

<h2> Multinomial Naive Bayes Classification <h2>

```python
# Building Models

cv = CountVectorizer()

mnb = MultinomialNB()

pipe = Pipeline([('cv', cv),('mnb', mnb)])

params = {
    'mnb__alpha':[0.1,0.25,0.5,0.75,1]
}

mnb_grid = GridSearchCV(pipe, params, cv=10)

# Training Models

mnb_grid.fit(X_train, y_train)
```

```python
mnb_grid.best_params_
```

```python
y_predict = mnb_grid.predict(X_test)

print('Accuracy score: ', format(accuracy_score(y_test, y_predict)))
print('Precision score: ', format(precision_score(y_test, y_predict)))
print('Recall score: ', format(recall_score(y_test, y_predict)))
print('F1 score: ', format(f1_score(y_test, y_predict)))
```

<h2> Logistic Regression <h2>

```python
# Building Models

cv = CountVectorizer()

lr = LogisticRegression()

pipe = Pipeline([('cv',cv),('lr',lr)])

params = {
    'lr__penalty':['l1','l2','elasticnet','none'],
    'lr__C':[0.25,0.5,0.75,1],
    'lr__class_weight':['balance',None],
    'lr__max_iter':[100,200,500]
}

lr_grid = GridSearchCV(pipe, params, cv=10)

# Traning Models

lr_grid.fit(X_train, y_train)
```

```python
lr_grid.best_params_
```

```python
y_predict = lr_grid.predict(X_test)

print('Accuracy score: ', format(accuracy_score(y_test, y_predict)))
print('Precision score: ', format(precision_score(y_test, y_predict)))
print('Recall score: ', format(recall_score(y_test, y_predict)))
print('F1 score: ', format(f1_score(y_test, y_predict)))
```
